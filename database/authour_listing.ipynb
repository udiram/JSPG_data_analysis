```python
import pandas as pd
```

```python
# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('JSPG author database.xlsx - Journal Article Database.csv')
df.head()
```

```python
# Step 2: Create a new column 'all_authors' that contains a list of authors for each article
def combine_authors(row):
    authors_list = []
    for i in range(1, 12):
        author = row[f'A{i} Email']
        if isinstance(author, str):
            authors_list.append(author.strip())
    return authors_list
```

```python
df['all_email'] = df.apply(combine_authors, axis=1)
```

```python
# Step 3: Delete the original 'author 1', 'author 2', and other author-related columns
# The axis parameter 1 indicates that we want to drop columns
df.drop(columns=[f'A{i} Email' for i in range(1, 13)], inplace=True)
```

```python
# Step 4: Delete the rows with a single element in the 'all_authors' list
df = df[df['all_email'].apply(lambda x: len(x) > 1)]
```

```python
# Step 5: Reset the DataFrame index to avoid gaps
df.reset_index(drop=True, inplace=True)
```
